```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import csv
```

```python
data = open("train.csv")

f = data.readlines()

count = 0
for i in f:
    count += 1
count-=1
print(count)
partition = 0.9
# print(partition*count)
X_train = np.zeros((int(count*partition),784), dtype=np.uint8)
Y_train = np.zeros((int(count*partition),1), dtype=np.int8)
X_validation = np.zeros((int((1-partition)*count),784), dtype=np.uint8)
Y_validation = np.zeros((int((1-partition)*count),1), dtype=np.int8)
for i in range(1,count):
    line = f[i].split(",")

    if i < partition*count:
        Y_train[i-1] = line[0]
        X_train[i-1] = line[1:]
    else:
        Y_validation[i-int(partition*count)-1] = line[0]
        X_validation[i-int(partition*count)-1] = line[1:]
```

```python
for i in range(10):
    print(Y_train[i])
    img = np.reshape(X_train[i],(28,28))
    plt.imshow(img,interpolation='nearest')
    plt.show()
```

```python
for i in range(10):
    print(Y_validation[i])
    img = np.reshape(X_validation[i],(28,28))
    plt.imshow(img)
    plt.show()
```

```python
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_validation = X_validation.reshape(X_validation.shape[0], 28, 28, 1)
```

```python
Y_train = tf.keras.utils.to_categorical(Y_train, 10)
Y_validation = tf.keras.utils.to_categorical(Y_validation, 10)
```

```python
Y_train.shape
```

```python
X_train = X_train/255
X_validation = X_validation/255
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(
        32, (3, 3), input_shape=(28, 28, 1), padding="same"
    ),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.Conv2D(
        64, (3, 3), input_shape=(28, 28, 32), padding="same"
    ),
    # Max-pooling layer, using 2x2 pool size
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
    tf.keras.layers.Dropout(0.25),
    tf.keras.layers.LeakyReLU(0.1),
    # Flatten units
    tf.keras.layers.Flatten(),

    # Add a hidden layer with dropout
    tf.keras.layers.Dense(128),
    tf.keras.layers.Dropout(0.5),

    # Add an output layer with output units for all 10 digits
    tf.keras.layers.Dense(10),
    tf.keras.layers.Activation("softmax")

])
```

```python
model.summary()
```

```python
model.compile(
    optimizer = "adam",
    loss = "categorical_crossentropy",
    metrics = ["accuracy"]
)
```

```python
model.fit(X_train, Y_train, epochs = 25)
```

```python
model.evaluate(X_validation, Y_validation, verbose = 2)
```
